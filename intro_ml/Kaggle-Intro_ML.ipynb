# Model Validation
Measure the performance of your model, so you can test and compare alternatives

### What is Model Validation
In most applications, the relevant measure of model quality is predictive accuracy.  

**Mean Absolute Error (MAE)**  
Prediction error  
- error = actual-predicted

With MAE metric, take abs value of each error, then take the average of those abs errors.  
"On average, our predictions are off by about X"

### Example

```python
import pandas as pd

# load data
melb_data = pd.read_csv('melb_data.csv')

# filter rows with missing price values
# dropna drops missing values (think of na as "not available")
filtered_melb_data = melb_data.dropna(axis=0)

filtered_melb_data.columns
```

```python
# Choose target and features
y = filtered_melb_data.Price

melb_features = ['Rooms', 'Bathroom', 'Landsize', 'BuildingArea', 'YearBuilt', 'Lattitude', 'Longtitude']
X = filtered_melb_data[melb_features]

from sklearn.tree import DecisionTreeRegressor

# define model
melb_model = DecisionTreeRegressor(random_state=1)

# fit model
melb_model.fit(X, y)
```

**Calculate MAE**

```python
from sklearn.metrics import mean_absolute_error

predicted_home_prices = melb_model.predict(X)
mean_absolute_error(y, predicted_home_prices)
```

### The Problem with "In-Sample Scores
Patterns are derived from the training set, so the model will appear accurate in the training data. You must test the model on new data (data that was not used to build the model) to get an accurate measure of prediction accuracy. Easiest way to do this is exculde part of the dataset and use as a validation data set.

```python
from sklearn.model_selection import train_test_split

# split data into training and validation data, for both features and target
# The split is based on a random number generator. Supplying a numeric value to
# the random_state argument guarantees we get the same split every tme we run this script.
train_X, val_X, train_y, val_y = train_test_split(X, y, random_state = 0)

# define model
melb_model = DecisionTreeRegressor()

# fit model
melb_model.fit(train_X, train_y)

# get predicted prices on validation data
val_predictions = melb_model.predict(val_X)
print(mean_absolute_error(val_y, val_predictions))
```

**Compare almost 500 dollar MAE to actual 250,000+ dollar MAE**

# Underfitting and Overfitting
## Experimenting with Different Models

**Overfitting**: where a mdoel matches the training data almost perfectly, but does poorly in validation and other new data. Too many features have been added in so that only a few data points are used to make a prediction.

**Underfitting**: when a model fails to capture important distinctions and patterns in the data, so it performs poorly even in training data.

### Example
*max_leaf_nodes* argument provides a sensible way to control overfitting vs underfitting. The more leaves we allow the model to make, the more we move from underfitting toward overfitting. We want a balance in the middle.

We can use a utility function to help compare MAE scores from different values for *max_leaf_nodes*:

```python
from sklearn.metrics import mean_absolute_error
from sklearn.tree import DecisionTreeRegressor

def get_mae(max_leaf_nodes, train_X, val_X, train_y, val_y):
    model = DecisionTreeRegressor(max_leaf_nodes=max_leaf_nodes, random_state=0)
    model.fit(train_X, train_y)
    preds_val = model.predict(val_X)
    mae = mean_absolute_error(val_y, preds_val)
    return(mae)

import pandas as pd

# load data
melb_data = pd.read_csv('melb_data.csv')

# filter rows with missing price values
# dropna drops missing values (think of na as "not available")
filtered_melb_data = melb_data.dropna(axis=0)

# Choose target and features
y = filtered_melb_data.Price

melb_features = ['Rooms', 'Bathroom', 'Landsize', 'BuildingArea', 'YearBuilt', 'Lattitude', 'Longtitude']
X = filtered_melb_data[melb_features]

from sklearn.model_selection import train_test_split

# split data into training and validation data, for both features and target
train_X, val_X, train_y, val_y = train_test_split(X, y, random_state = 0)

# use for-loop to compare the accuracy of models build with different values of max_leaf_nodes
for max_leaf_nodes in [5,50,500,5000]:
    my_mae = get_mae(max_leaf_nodes, train_X, val_X, train_y, val_y)
    print('Max leaf nodes: %d  \t\t Mean Absolute Error: %d' %(max_leaf_nodes, my_mae))
```

Here's the takeaway: Models can suffer from either:

- Overfitting: capturing spurious patterns that won't recur in the future, leading to less accurate predictions,  
or
- Underfitting: failing to capture relevant patterns, again leading to less accurate predictions.  

We use validation data, which isn't used in model training, to measure a candidate model's accuracy. This lets us try many candidate models and keep the best one.

# Exercise 4: Underfitting and Overfitting

Optimize the size of the tree in the model you created earlier to make better predictions.

```python
import pandas as pd
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_absolute_error
from sklearn.model_selection import train_test_split

# load data
home_data = pd.read_csv('./train.csv')

# create target object - y
y = home_data.SalePrice

# create feature list - X
features = ['LotArea', 'YearBuilt', '1stFlrSF', '2ndFlrSF', 'FullBath', 'BedroomAbvGr', 'TotRmsAbvGrd']
X = home_data[features]

# split dataset
train_X, val_X, train_y, val_y = train_test_split(X, y, random_state=1)

# specify model
iowa_model = DecisionTreeRegressor(random_state=1)

# fit model
iowa_model.fit(train_X, train_y)

# make validation predictions and calculate mean absolute error
val_predictions = iowa_model.predict(val_X)
val_mae = mean_absolute_error(val_predictions, val_y)
print('Validation MAE: {:,.0f}'.format(val_mae))
```

Write a loop that tries the following values for *max_leaf_nodes* from a set of possible values.

Call the *get_mae* funtion on each value of max_leaf_nodes. Store the output in some way that allows you to select the value of max_leaf_nodes that gives the most accurate model on your data

```python
candidate_max_leaf_nodes = [5,25,50,100,250,500]
#Write a loop to find the ideal tree size from candidate_max_leaf_nodes

for max_leaf_nodes in candidate_max_leaf_nodes:
    my_mae = get_mae(max_leaf_nodes, train_X, val_X, train_y, val_y)
    print('Max leaf nodes: %d  \t\t Mean Absolute Error: %d' %(max_leaf_nodes, my_mae))
```

# Random Forests
A deep decision tree with lots of leaves will overfit because each prediction is coming from historical data from only a few houses at its leaf. But a shallow tree with few leaves will perform poorly because it fails to capture as many distinctions in the raw data.

The random forest uses many trees, and it makes a prediction by averaging the predictions of each component tree. It generally has much better predictive accuracy than a single decision tree and it works well with default parameters. You can learn more models with better performance, but many are sensitive to getting the right parameters.

### Example
Load melbourne data and split

```python
import pandas as pd

# load data
melb_data = pd.read_csv('melb_data.csv')

# drop null values
melb_data = melb_data.dropna(axis=0)

# define target - y
y = melb_data.Price

# define features - X

melbourne_features = ['Rooms', 'Bathroom', 'Landsize', 'BuildingArea', 
                        'YearBuilt', 'Lattitude', 'Longtitude']
X = melb_data[melbourne_features]

from sklearn.model_selection import train_test_split

train_X, val_X, train_y, val_y = train_test_split(X,y, random_state=0)
```

Build random forest model similarly to a decision tree model. Use **RandomForestRegressor**

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error

forest_model = RandomForestRegressor(random_state=1)
forest_model.fit(train_X, train_y)
melb_preds = forest_model.predict(val_X)
print(mean_absolute_error(val_y, melb_preds))
```

# Exercise 5: Random Forest Model

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error

# load data
home_data = pd.read_csv('./train.csv')

# define target - y
y = home_data.SalePrice

# define features - X
features = ['LotArea', 'YearBuilt', '1stFlrSF', '2ndFlrSF', 'FullBath', 'BedroomAbvGr', 'TotRmsAbvGrd']
X = home_data[features]

# split the data
train_X, val_X, train_y, val_y = train_test_split(X,y, random_state=1)

# define model
rf_model = RandomForestRegressor(random_state=1)

# fit model
rf_model.fit(train_X, train_y)

# calculate MAE
rf_preds = rf_model.predict(val_X)
rf_val_mae = mean_absolute_error(val_y, rf_preds)

print("Validation MAE for Random Forest Model: {}".format(rf_val_mae))
```
